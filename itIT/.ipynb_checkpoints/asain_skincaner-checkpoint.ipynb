### 데이터 불러오기

```python
# 데이터 준비
df = pd.read_csv('C:/Users/gjaischool/CHU/itIT/heart_2020_cleaned.csv')
df
```

### 아시아인 데이터만 출력하기 라인번호 재설정
- 인종에서 Asian인만 빼서 설정

```python
df_asian = df[df['Race']=='Asian']
df_asian = df_asian.reset_index(drop=True)
df_asian
```

```python
df_asian.info()
```

### 범주형값을 수치값으로 변환(AgeCategory, Diabetic)
- OrfinalEncoder으로 라베링 실시
- AgeCategory, Diabetic값 OrdinalEncoder로 변환

```python
one_dr_encoder = OrdinalEncoder()
```

```python
age_df = df_asian[['AgeCategory']]
age_df
```

```python
age_df['AgeCategory'].value_counts()
```

```python
df_age = one_dr_encoder.fit_transform(age_df)
o_e_df= pd.DataFrame(df_age, columns=['Age'])
o_e_df.value_counts()
```

### Diabetic 값 라벨링

```python
df_diab= df_asian[['Diabetic']]
df_diab.value_counts()
```

```python
o_e_di = one_dr_encoder.fit_transform(df_diab)
di = pd.DataFrame(o_e_di, columns = ['Diabetic'])
di.value_counts()
```

```python
di = di.replace({'Diabetic': 1}, 0)
di = di.replace({'Diabetic': 2}, 1)
di = di.replace({'Diabetic': 3}, 1)
di
```

### Diabetic 값 제거

```python
X_df = df.drop(columns=['Diabetic'],axis=1)
X_df
```

### 변환했던 값 concat으로 합치기
- o_e_df : AgeCategory
- di : Diabetic

```python
X_df = pd.concat([X_df, o_e_df,di], axis=1, join='inner') 
X_df
```

### Age구간을 구분하여 컬럼 지정

```python
bins = [-1,1,3,5,7,13]
group_name = [1,2,3,4,5]
categorise = pd.cut(X_df['Age'],bins,labels=group_name)
X_df['Age']=categorise
X_df
```

### Age타입 변경 : int64, 불필요한 컬럼 drop ('AgeCategory','PhysicalHealth','MentalHealth','Race','GenHealth')

```python
X1_df = X_df.astype({'Age':'int64'})
```

```python
X1_df_1 = X1_df.drop(columns=['AlcoholDrinking','AgeCategory','PhysicalHealth','MentalHealth','Race','GenHealth'],axis=1)
X1_df_1
```

```python
X1_df_1['DiffWalking'] = [1 if each == 'Yes' else 0 for each in X1_df_1['DiffWalking']]
```

```python
X1_df_1['Asthma'] = [1 if each == 'Yes' else 0 for each in X1_df_1['Asthma']]
```

```python
X1_df_1 = X1_df_1.drop(columns=['SkinCancer'],axis=1)
X1_df_1
```

### 구간값을 실수값으로 구분하여 레이블 저장 및 타입 변경 : 원래 -object / 변경타입 - int
- 0 ~ 18.50 : 저체중 / groupnames = 1
- 18.50 ~ 24.99 : 정상 / groupnames = 2
- 25.00 ~ 29.99 : 과체중 / groupnames = 3
- 30.00 ~ 34.99 : 1단계 비만 / groupnames = 4
- 35.00 ~ 39.99 : 2단계 비만 / groupnames = 5
- 40 ~ : 고도비만 / groupnames = 6

```python
bins  = [0,18.50,24.99,29.99,34.99,39.99,99]
group_names = [1,2,3,4,5,6]   # bin a < 구간 <= b
categorise = pd.cut(X1_df_1['BMI'],bins,labels=group_names)
X1_df_1["BMI"]= categorise
X1_df_1 # 1: 저체중 2:정상 3:과체중 4: 1단계비만 5: 2단계 비만 6: 고도비만
```

```python
X1_df_1 = X1_df_1.astype({'BMI':'int64'})
```

```python
X1_df_1['BMI'].info()
```
