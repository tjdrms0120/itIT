```python
import pickle
import numpy as np

model = pickle.load(open('ML_model.pkl','rb'))

app = Flask(__name__)

@app.route('/', methods=['GET','POST'])
def main():
    return render_template('home.html')


@app.route('/predict', methods=['POST'])
def home():
    data1 = request.form['a']
    data2 = request.form['b']
    data3 = request.form['c']
    data4 = request.form['d']
    data5 = request.form['e']
    data6 = request.form['f']
    data7 = request.form['g']
    data8 = request.form['h']
    data9 = request.form['i']
    data10 = request.form['j']
    data11 = request.form['k']
    data12 = request.form['l']

    # 변수 값 지정
    data1 = int(data1)
    data2 = int(data2)
    data3 = int(data3)
    data4 = int(data4)
    data5 = int(data5)
    data6 = int(data6)
    data7 = int(data7)
    data8 = int(data8)
    data9 = int(data9)
    data10 = int(data10)
    data11 = int(data11)
    data12 = int(data12)

    final_features = np.array([(data1, data2, data3, data4, data5,
                   data6, data7, data8, data9, data10,
                   data11, data12)])
    sc=pickle.load(open('scaler.sav','rb'))
    final_features=sc.transform(final_features)
    prediction = model.predict(final_features)

    print(model)
    print(final_features)
    print(prediction)
    return render_template('after.html', data=prediction)

# host_addr = "192.168.0.2"
host_addr = "192.168.30.199"
port_num = "5000"
if __name__ =="__main__":
    app.run(host=host_addr, port=port_num)
```
